# subplots

```python
x=np.linspace(1,12,10)
y=x**2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b-*')
plt.subplot(1,2,2)
plt.plot(x,y,'r--')

```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'b*')
plt.subplot(2,2,2)
plt.plot(x,y,'r*')
plt.subplot(2,2,3)
plt.plot(x,y,'y*')
plt.subplot(2,2,4)
plt.plot(x,y,'g*')
```

```python
fig=plt.figure()
ax1=fig.add_axes([0.1,0.1,0.8,0.8])
ax2=fig.add_axes([0.2,0.5,0.43,0.3])

#[left ,bottom, width, height]

ax1.plot(x,y)
ax2.plot(y,x)
```

# pie chart

```python
import matplotlib.pyplot as plt
labels=["a","b","c","d"]
numbers=[12,13,21,14]
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.pie(numbers,labels=labels,autopct="%0.2f%%")
```

```python
import matplotlib.pyplot as plt
labels=["a","b","c","d"]
numbers=[12,13,21,14]
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.pie(numbers,labels=labels,autopct="%0.2f%%",colors=["red","blue","orange","green"])
```

```python
import matplotlib.pyplot as plt
labels=["a","b","c","d"]
numbers=[12,13,21,14]
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.pie(numbers,labels=labels,autopct="%0.2f%%",colors=["red","blue","orange","green"])
axes1.legend()
```
